```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from collections import Counter

# --- CÁC IMPORT TỪ PROJECT CỦA BẠN ---
import config
from utils import load_trained_model
from data_operations.data_preprocessing import import_cmmd_dataset

# =================================================================
# BƯỚC 1: CẤU HÌNH VÀ TẢI LẠI MÔ HÌNH CMMD ĐÃ HUẤN LUYỆN
# =================================================================
print("--- 1. Tải lại mô hình CMMD đã huấn luyện ---")

# --- Đặt các thông số để tải đúng model ---
# LƯU Ý: Hãy thay thế bằng tên chính xác của file model CMMD bạn đã lưu
# Tên này phải khớp với tên bạn đã đặt trong lúc huấn luyện
MODEL_NAME_CMMD = 'CMMD_ClassWeights_v2' # Ví dụ, thay bằng tên đúng
MODEL_PATH_CMMD = f'/kaggle/working/BreastMammo/saved_models/dataset-CMMD_type-all_model-MobileNet_lr-0.001_b-16_e1-25_e2-40_roi-False_{MODEL_NAME_CMMD}.weights.h5'

# Tải lại model
try:
    from cnn_models.cnn_model import CnnModel
    cnn_cmmd = CnnModel("MobileNet", num_classes=2)
    cnn_cmmd.model.load_weights(MODEL_PATH_CMMD)
    model_cmmd = cnn_cmmd.model
    model_cmmd.summary()
    print(f"Tải thành công model CMMD từ: {MODEL_PATH_CMMD}")

except Exception as e:
    print(f"Lỗi khi tải model CMMD: {e}")
    model_cmmd = None

# =================================================================
# BƯỚC 2: TẢI VÀ CHUẨN BỊ DỮ LIỆU TEST CỦA CMMD
# =================================================================
if model_cmmd:
    print("\n--- 2. Chuẩn bị dữ liệu CMMD để visualize ---")
    DATA_DIR_CMMD = '/kaggle/input/breastdata/CMMD/CMMD'
    le_cmmd = LabelEncoder()
    
    # Tải lại toàn bộ dữ liệu CMMD (chỉ cần tập test)
    X_loaded_cmmd, y_numeric_labels_cmmd = import_cmmd_dataset(
        data_dir=DATA_DIR_CMMD,
        label_encoder=le_cmmd,
    )
    
    le_cmmd.fit(np.array(['Benign', 'Malignant'])) # Fit với các lớp đã biết
    
    # Chia lại y hệt như lúc train để có tập test chính xác
    _, X_test_cmmd, _, y_test_numeric_cmmd = train_test_split(
        X_loaded_cmmd, y_numeric_labels_cmmd, test_size=0.2, stratify=y_numeric_labels_cmmd, random_state=42
    )
    print(f"Đã tải và chia xong, có {len(X_test_cmmd)} ảnh trong tập test CMMD.")

# =================================================================
# BƯỚC 3: HÀM GRAD-CAM (Không cần định nghĩa lại)
# =================================================================
# Giả sử các hàm make_gradcam_heatmap và display_gradcam đã được định nghĩa ở cell trước
# Nếu chưa, bạn cần copy chúng vào đây.
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

def display_gradcam(img, heatmap, alpha=0.4):
    img_u8 = (img * 255).astype(np.uint8)
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    superimposed_img = cv2.addWeighted(cv2.cvtColor(img_u8, cv2.COLOR_RGB2BGR), 1-alpha, heatmap, alpha, 0)
    return cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB)

# =================================================================
# BƯỚC 4: TẠO VÀ HIỂN THỊ VISUALIZATION CHO CMMD
# =================================================================
if model_cmmd and 'X_test_cmmd' in locals():
    print("\n--- 4. Tạo và hiển thị Grad-CAM cho CMMD ---")
    
    # Tên của lớp conv cuối cùng trong MobileNetV2
    LAST_CONV_LAYER_NAME_CMMD = "out_relu"

    # Chọn một vài ảnh từ tập test của CMMD để visualize
    benign_indices_cmmd = np.where(y_test_numeric_cmmd == 0)[0]
    malignant_indices_cmmd = np.where(y_test_numeric_cmmd == 1)[0]
    
    sample_indices_cmmd = np.concatenate([
        np.random.choice(benign_indices_cmmd, 2, replace=False),
        np.random.choice(malignant_indices_cmmd, 2, replace=False)
    ])

    plt.figure(figsize=(12, 12))
    
    for i, img_idx in enumerate(sample_indices_cmmd):
        # Lấy ảnh từ tập test CMMD
        original_img_cmmd = X_test_cmmd[img_idx]
        
        # Xử lý kênh cho model (CMMD là ảnh 1 kênh, MobileNet cần 3 kênh)
        if original_img_cmmd.shape[-1] == 1:
            img_to_predict = cv2.cvtColor(original_img_cmmd, cv2.COLOR_GRAY2RGB)
        else:
            img_to_predict = original_img_cmmd
            
        img_array = np.expand_dims(img_to_predict, axis=0)

        # Dự đoán để lấy nhãn
        preds = model_cmmd.predict(img_array)
        predicted_class_idx = np.argmax(preds[0])
        predicted_class_name = le_cmmd.classes_[predicted_class_idx]
        true_class_name = le_cmmd.classes_[y_test_numeric_cmmd[img_idx]]

        # Tạo heatmap
        heatmap = make_gradcam_heatmap(img_array, model_cmmd, LAST_CONV_LAYER_NAME_CMMD)
        
        # Overlay heatmap lên ảnh
        superimposed_img = display_gradcam(img_to_predict, heatmap, alpha=0.5)

        # Hiển thị
        plt.subplot(4, 2, 2*i + 1)
        plt.imshow(img_to_predict)
        plt.title(f"Original CMMD - True: {true_class_name}")
        plt.axis('off')

        plt.subplot(4, 2, 2*i + 2)
        plt.imshow(superimposed_img)
        plt.title(f"Grad-CAM - Predicted: {predicted_class_name}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```
