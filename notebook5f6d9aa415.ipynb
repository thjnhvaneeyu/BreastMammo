```python
import numpy as np
import cv2
from scipy.ndimage import gaussian_filter
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

def preprocess_and_visualize_cmmd_with_clahe(image_path):
    """
    Tải, tiền xử lý và trực quan hóa hình ảnh CMMD ở mỗi bước, bao gồm cả CLAHE.

    Các bước xử lý bao gồm:
    1. Chuẩn hóa (Normalization)
    2. Lọc Gaussian (Gaussian filtering)
    3. Áp dụng CLAHE để tăng cường độ tương phản
    4. Thay đổi kích thước và thêm đệm (Resize and padding)
    """
    # Tải ảnh gốc bằng OpenCV và chuyển sang ảnh xám
    original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if original_image is None:
        print(f"Lỗi: Không thể tải hình ảnh từ đường dẫn: {image_path}")
        return

    image_array = np.array(original_image)

    # Tạo một figure để hiển thị tất cả các bước
    plt.figure(figsize=(24, 4)) # Tăng kích thước để chứa thêm 1 ảnh

    # --- Hiển thị ảnh gốc ---
    plt.subplot(1, 6, 1)
    plt.imshow(image_array, cmap='gray')
    plt.title('1. Ảnh Gốc')
    plt.axis('off')

    # 1. Chuẩn hóa (Normalization): scale giá trị pixel về khoảng [0, 1]
    normalized_image = (image_array - np.min(image_array)) / (np.max(image_array) - np.min(image_array) + 1e-8)

    plt.subplot(1, 6, 2)
    plt.imshow(normalized_image, cmap='gray')
    plt.title('2. Đã Chuẩn Hóa')
    plt.axis('off')

    # 2. Lọc Gaussian (Gaussian filtering): giảm nhiễu
    # (sigma=0.5 là một giá trị hợp lý, tương tự như trong script cmmdconvert.py của bạn)
    smoothed_image = gaussian_filter(normalized_image, sigma=0.5)

    plt.subplot(1, 6, 3)
    plt.imshow(smoothed_image, cmap='gray')
    plt.title('3. Lọc Gaussian')
    plt.axis('off')

    # --- BƯỚC MỚI: ÁP DỤNG CLAHE ---
    # Chuyển ảnh float [0,1] sang uint8 [0,255] để dùng CLAHE
    image_uint8_for_clahe = (smoothed_image * 255).astype(np.uint8)
    
    # Khởi tạo CLAHE với các tham số từ config.py của bạn
    clahe_processor = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    image_after_clahe = clahe_processor.apply(image_uint8_for_clahe)
    
    # Chuyển lại ảnh float [0,1] để xử lý tiếp
    image_clahe_float = image_after_clahe.astype(np.float32) / 255.0

    plt.subplot(1, 6, 4)
    plt.imshow(image_clahe_float, cmap='gray')
    plt.title('4. Áp dụng CLAHE')
    plt.axis('off')

    # 3. Chuyển đổi về uint8 để có thể xử lý bằng PIL
    # Sử dụng ảnh đã qua CLAHE
    uint8_image_after_clahe = (image_clahe_float * 255).astype(np.uint8)

    # 4. Thay đổi kích thước về 224x224 và thêm đệm để giữ tỷ lệ khung hình
    desired_size = 224
    image_pil = Image.fromarray(uint8_image_after_clahe)
    old_size = image_pil.size  # (width, height)
    ratio = float(desired_size) / max(old_size)
    new_size = tuple([int(x * ratio) for x in old_size])

    image_pil_resized = image_pil.resize(new_size, Image.LANCZOS)

    delta_w = desired_size - new_size[0]
    delta_h = desired_size - new_size[1]
    padding = (delta_w // 2, delta_h // 2, delta_w - (delta_w // 2), delta_h - (delta_h // 2))
    padded_image = ImageOps.expand(image_pil_resized, padding, fill=0)  # Đệm màu đen

    plt.subplot(1, 6, 5)
    plt.imshow(padded_image, cmap='gray')
    plt.title('5. Thêm Đệm & Resize')
    plt.axis('off')

    # Chuyển đổi lại thành mảng numpy
    final_image_array = np.array(padded_image)

    # Đảm bảo kích thước cuối cùng chính xác là 224x224
    if final_image_array.shape[:2] != (224, 224):
        final_image = cv2.resize(final_image_array, (224, 224))
    else:
        final_image = final_image_array

    print("Kích thước ảnh đầu ra:", final_image.shape)

    plt.subplot(1, 6, 6)
    plt.imshow(final_image, cmap='gray')
    plt.title('6. Ảnh Cuối Cùng (224x224)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

    return final_image

# --- CÁCH SỬ DỤNG ---
# Thay 'path/to/your/mammogram.png' bằng đường dẫn đến file ảnh nhũ ảnh của bạn.
# Ví dụ sử dụng lại đường dẫn ảnh CMMD bạn đã cung cấp:
# Lưu ý: bạn có thể cần cung cấp đường dẫn đầy đủ và chính xác trên máy của mình.
try:
    image_path = '/kaggle/input/breastdata/CMMD/CMMD/malignant/D1-0046_1-2.png'
    final_processed_image = preprocess_and_visualize_cmmd_with_clahe(image_path)
except FileNotFoundError:
    print(f"Lỗi: Không tìm thấy file ảnh tại '{image_path}'. Vui lòng cập nhật đường dẫn chính xác.")
    # Fallback path cho môi trường local (ví dụ)
    # local_path = "path/to/your/local/image.png"
    # final_processed_image = preprocess_and_visualize_cmmd_with_clahe(local_path)
```

```python
import os
import re
import cv2
import pydicom
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydicom.errors import InvalidDicomError

# --- CÁC HÀM TIỆN ÍCH (Lấy từ logic trong mã nguồn của bạn) ---

def load_roi_coordinates(roi_path: str) -> list:
    """Tải tọa độ từ một file .roi bất kỳ."""
    coords = []
    if not os.path.exists(roi_path):
        # print(f"Cảnh báo: Không tìm thấy file ROI tại {roi_path}")
        return coords
    try:
        with open(roi_path, 'r', encoding='utf-8', errors='ignore') as f:
            raw_content = f.read()
        # Regex để tìm các cặp tọa độ trong dấu ngoặc {} hoặc ()
        pts_matches = re.findall(r'[\{\(]\s*([\d\.]+)\s*,\s*([\d\.]+)\s*[\)\}]', raw_content)
        for xs, ys in pts_matches:
            # Bỏ qua các cặp tọa độ mặc định không có ý nghĩa
            if abs(float(xs) - 75.0) < 1e-6 and abs(float(ys) - 19.0) < 1e-6:
                continue
            coords.append((int(float(xs)), int(float(ys))))
    except Exception as e:
        print(f"Lỗi khi đọc file ROI {roi_path}: {e}")
    return coords

def remove_pectoral_muscle(image_array: np.ndarray, pectoral_coords: list) -> np.ndarray:
    """Xóa vùng cơ ngực bằng cách tô đen vùng mask."""
    if not pectoral_coords or len(pectoral_coords) < 3:
        return image_array.copy() # Trả về ảnh gốc nếu không có tọa độ
    
    mask = np.ones(image_array.shape[:2], dtype=np.uint8)
    points = np.array(pectoral_coords, dtype=np.int32)
    
    try:
        # Tô đen vùng đa giác được định nghĩa bởi các điểm tọa độ
        cv2.fillPoly(mask, [points], 0)
        image_cleaned = image_array.copy()
        # Áp dụng mask lên ảnh (chỉ hoạt động với ảnh thang độ xám 2D)
        image_cleaned[mask == 0] = 0.0
        return image_cleaned
    except Exception as e:
        print(f"Lỗi khi áp dụng fillPoly để xóa cơ ngực: {e}")
        return image_array.copy()

# --- HÀM TRỰC QUAN HÓA CHÍNH ---

def visualize_inbreast_processing(dicom_path, lesion_roi_path, pectoral_roi_path):
    """
    Tải, tiền xử lý và trực quan hóa ảnh INbreast ở mỗi bước.
    """
    # --- Tải ảnh DICOM ---
    try:
        dicom_data = pydicom.dcmread(dicom_path)
        image_array = dicom_data.pixel_array.astype(np.float32)
    except (InvalidDicomError, FileNotFoundError) as e:
        print(f"Lỗi: Không thể tải hoặc đọc file DICOM: {dicom_path}\n{e}")
        return

    # Tạo một figure để hiển thị tất cả các bước
    plt.figure(figsize=(24, 4)) # Tăng kích thước để dễ nhìn

    # --- 1. Hiển thị ảnh gốc ---
    plt.subplot(1, 6, 1)
    plt.imshow(image_array, cmap='gray')
    plt.title('1. Ảnh DICOM Gốc')
    plt.axis('off')

    # --- 2. Chuẩn hóa cường độ pixel về [0, 1] ---
    min_val, max_val = np.min(image_array), np.max(image_array)
    normalized_image = (image_array - min_val) / (max_val - min_val + 1e-8)
    normalized_image = np.clip(normalized_image, 0.0, 1.0)
    
    plt.subplot(1, 6, 2)
    plt.imshow(normalized_image, cmap='gray')
    plt.title('2. Đã Chuẩn Hóa')
    plt.axis('off')

    # --- 3. Loại bỏ cơ ngực ---
    pectoral_coords = load_roi_coordinates(pectoral_roi_path)
    image_no_pectoral = remove_pectoral_muscle(normalized_image, pectoral_coords)

    plt.subplot(1, 6, 3)
    plt.imshow(image_no_pectoral, cmap='gray')
    plt.title('3. Đã Xóa Cơ Ngực')
    plt.axis('off')

    # --- 4. Áp dụng CLAHE ---
    image_uint8 = (image_no_pectoral * 255).astype(np.uint8)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    image_clahe = clahe.apply(image_uint8).astype(np.float32) / 255.0

    plt.subplot(1, 6, 4)
    plt.imshow(image_clahe, cmap='gray')
    plt.title('4. Sau Khi Dùng CLAHE')
    plt.axis('off')
    
    # --- 5. Cắt vùng ROI của tổn thương ---
    lesion_coords = load_roi_coordinates(lesion_roi_path)
    if not lesion_coords:
        print(f"Không tìm thấy tọa độ tổn thương trong file: {lesion_roi_path}")
        plt.tight_layout()
        plt.show()
        return

    xs = [p[0] for p in lesion_coords]; ys = [p[1] for p in lesion_coords]
    x_min, x_max = int(min(xs)), int(max(xs))
    y_min, y_max = int(min(ys)), int(max(ys))
    
    # Cắt ROI từ ảnh đã qua CLAHE
    lesion_roi_cropped = image_clahe[y_min:y_max+1, x_min:x_max+1]

    plt.subplot(1, 6, 5)
    plt.imshow(lesion_roi_cropped, cmap='gray')
    plt.title('5. Cắt ROI Tổn Thương')
    plt.axis('off')

    # --- 6. Resize ROI về kích thước cuối cùng ---
    target_size = (224, 224) # Kích thước mục tiêu cho mô hình
    final_roi = cv2.resize(lesion_roi_cropped, target_size, interpolation=cv2.INTER_AREA)
    
    plt.subplot(1, 6, 6)
    plt.imshow(final_roi, cmap='gray')
    plt.title(f'6. ROI Cuối Cùng ({target_size[0]}x{target_size[1]})')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

# --- VÍ DỤ SỬ DỤNG ---
if __name__ == '__main__':
    # THAY ĐỔI CÁC ĐƯỜNG DẪN NÀY CHO PHÙ HỢP VỚI MÁY CỦA BẠN
    
    # Đường dẫn gốc tới dữ liệu INbreast
    INBREAST_DATA_DIR = "/kaggle/input/breastdata/INbreast/INbreast"
    
    # Đường dẫn tới thư mục chứa ROI của cơ ngực
    PECTORAL_DIR = "/kaggle/input/breastdata/INbreast/INbreast/PectoralMuscle/Pectoral Muscle ROI" 

    # Chọn một ID ảnh để thử nghiệm, dựa trên ví dụ của bạn
    IMAGE_ID = "20586986" 

    # Giả định tên file ROI tổn thương có dạng <ID>_1.roi hoặc <ID>.roi
    # và file cơ ngực có dạng <ID>_muscle.roi
    # Bạn có thể cần điều chỉnh logic tìm file nếu tên file khác
    dicom_file_path = ""
    # Tìm file DICOM có tên bắt đầu bằng IMAGE_ID
    for f in os.listdir(os.path.join(INBREAST_DATA_DIR, "AllDICOMs")):
        if f.startswith(IMAGE_ID) and f.lower().endswith(".dcm"):
            dicom_file_path = os.path.join(INBREAST_DATA_DIR, "AllDICOMs", f)
            break
            
    lesion_roi_file_path = os.path.join(INBREAST_DATA_DIR, "AllROI", f"{IMAGE_ID}_1.roi") # Thử với `_1`
    if not os.path.exists(lesion_roi_file_path):
        lesion_roi_file_path = os.path.join(INBREAST_DATA_DIR, "AllROI", f"{IMAGE_ID}.roi") # Thử không có `_1`

    pectoral_roi_file_path = os.path.join(PECTORAL_DIR, f"{IMAGE_ID}_muscle.roi")
    
    print(f"Trực quan hóa pipeline cho Ảnh ID: {IMAGE_ID}")
    print(f" - Đường dẫn DICOM: {dicom_file_path}")
    print(f" - Đường dẫn ROI tổn thương: {lesion_roi_file_path}")
    print(f" - Đường dẫn ROI cơ ngực: {pectoral_roi_file_path}")
    
    # Kiểm tra sự tồn tại của các file trước khi chạy
    if not dicom_file_path or not os.path.exists(dicom_file_path):
        print(f"\nLỖI: Không tìm thấy file DICOM cho ID {IMAGE_ID}. Vui lòng kiểm tra lại.")
    elif not os.path.exists(lesion_roi_file_path):
        print(f"\nLỖI: Không tìm thấy file ROI của tổn thương. Vui lòng kiểm tra lại.")
    else:
        # Chạy hàm trực quan hóa
        visualize_inbreast_processing(
            dicom_path=dicom_file_path,
            lesion_roi_path=lesion_roi_file_path,
            pectoral_roi_path=pectoral_roi_file_path
        )
```

```python
import os
import re
import cv2
import pydicom
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydicom.errors import InvalidDicomError

# --- CÁC HÀM TIỀN XỬ LÝ (Tương tự phiên bản ROI) ---

def load_roi_coordinates(roi_path: str) -> list:
    """Tải tọa độ từ một file .roi bất kỳ."""
    coords = []
    if not os.path.exists(roi_path):
        return coords
    try:
        with open(roi_path, 'r', encoding='utf-8', errors='ignore') as f:
            raw_content = f.read()
        pts_matches = re.findall(r'[\{\(]\s*([\d\.]+)\s*,\s*([\d\.]+)\s*[\)\}]', raw_content)
        for xs, ys in pts_matches:
            if abs(float(xs) - 75.0) < 1e-6 and abs(float(ys) - 19.0) < 1e-6:
                continue
            coords.append((int(float(xs)), int(float(ys))))
    except Exception as e:
        print(f"Lỗi khi đọc file ROI {roi_path}: {e}")
    return coords

def remove_pectoral_muscle(image_array: np.ndarray, pectoral_coords: list) -> np.ndarray:
    """Xóa vùng cơ ngực bằng cách tô đen vùng mask."""
    if not pectoral_coords or len(pectoral_coords) < 3:
        return image_array.copy()
    mask = np.ones(image_array.shape[:2], dtype=np.uint8)
    points = np.array(pectoral_coords, dtype=np.int32)
    try:
        cv2.fillPoly(mask, [points], 0)
        image_cleaned = image_array.copy()
        image_cleaned[mask == 0] = 0.0
        return image_cleaned
    except Exception as e:
        print(f"Lỗi khi áp dụng fillPoly để xóa cơ ngực: {e}")
        return image_array.copy()

# --- HÀM TẢI DỮ LIỆU CHÍNH CHO PHIÊN BẢN FULL IMAGE ---

def load_inbreast_full_image_version(data_root_dir: str, pectoral_roi_dir: str, target_size=(224, 224)):
    """
    Tải và xử lý dữ liệu INbreast, sử dụng toàn bộ ảnh.
    """
    dicom_dir = os.path.join(data_root_dir, "AllDICOMs")
    csv_path = os.path.join(data_root_dir, "INbreast.csv")

    if not os.path.exists(csv_path):
        print(f"Lỗi: không tìm thấy file {csv_path}")
        return [], []

    df = pd.read_csv(csv_path, sep=';')
    df.columns = [c.strip() for c in df.columns]
    birad_map = {str(row['File Name']).strip().split('_')[0]: str(row['Bi-Rads']).strip() for _, row in df.iterrows()}

    X_processed = []
    y_labels = []

    print(f"Bắt đầu quét thư mục DICOM: {dicom_dir}")
    for dicom_file in sorted(os.listdir(dicom_dir)):
        if not dicom_file.lower().endswith(".dcm"):
            continue

        # 1. Trích xuất ID từ tên file DICOM
        patient_id = dicom_file.split('_')[0]
        
        # 2. Tìm file ROI cơ ngực tương ứng
        pectoral_roi_path = os.path.join(pectoral_roi_dir, f"{patient_id}_muscle.roi")
        
        # 3. Lấy nhãn
        birad_value = birad_map.get(patient_id)
        if birad_value is None or birad_value in ['1', '3']:
            continue
        label = "Malignant" if birad_value in ['4a', '4b', '4c', '5', '6'] else "Benign"
        
        # 4. Tải và xử lý ảnh
        try:
            dicom_path = os.path.join(dicom_dir, dicom_file)
            dicom_image = pydicom.dcmread(dicom_path).pixel_array.astype(np.float32)
            
            # Chuẩn hóa
            normalized_image = (dicom_image - np.min(dicom_image)) / (np.max(dicom_image) - np.min(dicom_image) + 1e-8)
            
            # Xóa cơ ngực
            pectoral_coords = load_roi_coordinates(pectoral_roi_path)
            image_no_pectoral = remove_pectoral_muscle(normalized_image, pectoral_coords)
            
            # Áp dụng CLAHE
            image_uint8 = (image_no_pectoral * 255).astype(np.uint8)
            clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
            image_clahe = clahe.apply(image_uint8).astype(np.float32) / 255.0
            
            # Resize toàn bộ ảnh về kích thước mục tiêu
            final_image = cv2.resize(image_clahe, target_size, interpolation=cv2.INTER_AREA)

            X_processed.append(final_image)
            y_labels.append(label)
        
        except Exception as e:
            print(f"Lỗi xử lý file ID {patient_id}: {e}")
            continue
            
    return np.array(X_processed), np.array(y_labels)

# --- VÍ DỤ SỬ DỤNG PHIÊN BẢN FULL IMAGE ---
if __name__ == '__main__':
    DATA_ROOT = "/kaggle/input/breastdata/INbreast/INbreast"
    PECTORAL_DIR = "/kaggle/input/breastdata/INbreast/INbreast/PectoralMuscle/Pectoral Muscle ROI"
    
    print("\n--- Đang chạy phiên bản xử lý Toàn bộ ảnh ---")
    X_full, y_full = load_inbreast_full_image_version(DATA_ROOT, PECTORAL_DIR)
    
    print(f"Hoàn thành! Đã tải và xử lý {len(X_full)} ảnh.")
    if len(X_full) > 0:
        print(f"Shape của mảng ảnh X: {X_full.shape}")
        print(f"Shape của mảng nhãn y: {y_full.shape}")
        # Trực quan hóa một vài mẫu
        plt.figure(figsize=(12, 4))
        for i in range(min(5, len(X_full))):
            plt.subplot(1, 5, i + 1)
            plt.imshow(X_full[i], cmap='gray')
            plt.title(y_full[i])
            plt.axis('off')
        plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from collections import Counter

# --- CÁC IMPORT TỪ PROJECT CỦA BẠN ---
import config
from utils import load_trained_model
# Giả sử hàm import_cmmd_dataset và các hàm khác nằm ở đây
from data_operations.data_preprocessing import import_cmmd_dataset 
from cnn_models.mobilenet_v2 import create_mobilenet_model # Import hàm tạo model của bạn

# =================================================================
# BƯỚC 1: CẤU HÌNH VÀ TẢI LẠI MÔ HÌNH CMMD ĐÃ HUẤN LUYỆN
# =================================================================
print("--- 1. Tải lại mô hình CMMD đã huấn luyện ---")

# --- Đặt các thông số để tải đúng model ---
# Tên này phải khớp với tên bạn đã đặt trong lúc huấn luyện
MODEL_NAME_CMMD = 'CMMD_Oversampling_Balanced' # Ví dụ, thay bằng tên đúng
# Đường dẫn này có thể cần điều chỉnh lại cho đúng với tên model và các tham số
MODEL_PATH_CMMD = f'/kaggle/working/BreastMammo/saved_models/dataset-CMMD_type-all_model-MobileNet_lr-0.001_b-16_e1-50_e2-50_roi-False_CMMD_MobileNet_ReducedParams_Elastic2.h5'
# Khởi tạo lại kiến trúc model rỗng
# Cần set config để hàm create_mobilenet_model biết đang làm việc với CMMD
config.dataset = "CMMD"
config.model = "MobileNet"
config.verbose_mode = True

try:
    # Tạo lại kiến trúc rỗng
    model_cmmd = create_mobilenet_model(num_classes=2)
    # Tải trọng số đã huấn luyện vào kiến trúc này
    model_cmmd.load_weights(MODEL_PATH_CMMD)
    print(f"\n--- Tải thành công model CMMD từ: {MODEL_PATH_CMMD} ---")
    model_cmmd.summary()

except Exception as e:
    print(f"\nLỗi khi tải model CMMD: {e}")
    model_cmmd = None

# =================================================================
# BƯỚC 2: TẢI VÀ CHUẨN BỊ DỮ LIỆU TEST CỦA CMMD
# =================================================================
if model_cmmd:
    print("\n--- 2. Chuẩn bị dữ liệu CMMD để visualize ---")
    DATA_DIR_CMMD = '/kaggle/input/breastdata/CMMD/CMMD'
    le_cmmd = LabelEncoder()
    le_cmmd.fit(['Benign', 'Malignant'])
    
    X_loaded_cmmd, y_numeric_labels_cmmd = import_cmmd_dataset(
        data_dir=DATA_DIR_CMMD,
        label_encoder=le_cmmd,
    )
    
    # Chia lại y hệt như lúc train để có tập test chính xác
    _, X_test_cmmd, _, y_test_numeric_cmmd = train_test_split(
        X_loaded_cmmd, y_numeric_labels_cmmd, test_size=0.2, stratify=y_numeric_labels_cmmd, random_state=42
    )
    print(f"Đã tải và chia xong, có {len(X_test_cmmd)} ảnh trong tập test CMMD.")

# =================================================================
# =================================================================
# BƯỚC 3: ĐỊNH NGHĨA HÀM GRAD-CAM (Phiên bản nâng cấp)
# =================================================================

# --- Hàm make_gradcam_heatmap giữ nguyên, không cần sửa ---
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    # ... (giữ nguyên code của hàm này)
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()


# --- THAY THẾ HÀM display_gradcam CŨ BẰNG HÀM MỚI NÀY ---
def display_gradcam(original_img, heatmap, alpha=0.6, colormap=cv2.COLORMAP_JET):
    """
    Hòa trộn heatmap lên ảnh gốc một cách "nghệ thuật" hơn.
    :param original_img: Ảnh gốc (dạng float, 3 kênh, giá trị [0, 1]).
    :param heatmap: Heatmap từ Grad-CAM (dạng float 2D, giá trị [0, 1]).
    :param alpha: Độ trong suốt của heatmap.
    :param colormap: Bảng màu của OpenCV để tô màu cho heatmap.
    :return: Ảnh đã được phủ heatmap.
    """
    # 1. Chuyển heatmap sang dạng ảnh màu
    heatmap_resized = cv2.resize(heatmap, (original_img.shape[1], original_img.shape[0]))
    heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap_resized), colormap)
    heatmap_color_rgb = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    heatmap_float = heatmap_color_rgb.astype(np.float32) / 255.0

    # 2. Chuyển ảnh gốc sang float32 nếu chưa phải
    img_float = original_img.astype(np.float32)

    # 3. Hòa trộn (blending)
    # Công thức: blended = (1 - alpha) * original + alpha * heatmap
    blended_img = (1 - alpha) * img_float + alpha * heatmap_float
    
    # 4. Tạo một mask từ heatmap để chỉ áp dụng hiệu ứng ở những vùng "nóng"
    # Điều này giữ cho các vùng nền đen không bị ám màu
    heatmap_mask = np.expand_dims(heatmap_resized, axis=-1)
    
    # 5. Kết hợp ảnh gốc và ảnh đã hòa trộn dựa trên mask
    # Vùng nào có heatmap > 0 thì dùng ảnh blended, còn lại dùng ảnh gốc
    superimposed_img = (img_float * (1 - heatmap_mask) + blended_img * heatmap_mask)
    
    # 6. Clip lại giá trị để đảm bảo nằm trong khoảng [0, 1] và trả về
    return np.clip(superimposed_img, 0, 1)

# =================================================================
# BƯỚC 4: TẠO VÀ HIỂN THỊ VISUALIZATION CHO CNN
# =================================================================
if model_cnn and 'X_test_cnn' in locals():
    print("\n--- 4. Tạo và hiển thị Grad-CAM cho mô hình CNN ---")
    
    # **QUAN TRỌNG**: Tên của lớp conv cuối cùng trong mô hình của bạn
    LAST_CONV_LAYER_NAME_CNN = "Conv2" 

    # Chọn ảnh mẫu
    sample_indices = np.random.choice(len(X_test_cnn), 4, replace=False)

    plt.figure(figsize=(12, 12))
    
    for i, img_idx in enumerate(sample_indices):
        original_img = X_test_cnn[img_idx]
        
        # Mô hình CNN của bạn nhận đầu vào 1 kênh
        if original_img.shape[-1] != 1:
            print(f"Ảnh đầu vào có {original_img.shape[-1]} kênh, đang chuyển về 1 kênh.")
            original_img = cv2.cvtColor(original_img, cv2.COLOR_RGB2GRAY)
            original_img = np.expand_dims(original_img, axis=-1)

        img_array = np.expand_dims(original_img, axis=0)

        # Dự đoán
        preds = model_cnn.predict(img_array)
        predicted_class_idx = np.argmax(preds[0])
        predicted_class_name = le.classes_[predicted_class_idx]
        true_class_name = le.classes_[y_test_numeric_cnn[img_idx]]

        # Tạo heatmap
        heatmap = make_gradcam_heatmap(img_array, model_cnn, LAST_CONV_LAYER_NAME_CNN)
        
        # Chuyển ảnh xám sang RGB để phủ màu heatmap
        img_to_display = cv2.cvtColor(original_img, cv2.COLOR_GRAY2RGB)
        superimposed_img = display_gradcam(img_to_display, heatmap, alpha=0.5)

        # Hiển thị
        plt.subplot(4, 2, 2*i + 1)
        plt.imshow(original_img, cmap='gray') # Hiển thị ảnh gốc dạng xám
        plt.title(f"Original CNN Input - True: {true_class_name}")
        plt.axis('off')

        plt.subplot(4, 2, 2*i + 2)
        plt.imshow(superimposed_img)
        plt.title(f"Grad-CAM - Predicted: {predicted_class_name}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
import os  # ✅ Thêm thư viện os để làm việc với đường dẫn file
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# --- CÁC IMPORT TỪ PROJECT CỦA BẠN ---
import config
from cnn_models.basic_cnn import create_basic_cnn_model

# =================================================================
# BƯỚC 1: CẤU HÌNH VÀ TẢI LẠI MÔ HÌNH (ĐÃ SỬA LỖI BUILD)
# =================================================================
print("--- 1. Tải lại mô hình CNN đã huấn luyện cho CMMD ---")

MODEL_PATH_CMMD_CNN = f'/kaggle/input/cnnelastic1/tensorflow2/default/1/dataset-CMMD_type-all_model-CNN_lr-0.001_b-16_e1-50_e2-50_roi-False_CMMD_CNN_ReducedParams_Elastic1.weights.h5'

model_cnn_cmmd_built = None # Model cuối cùng sẽ được sử dụng

try:
    # 1. Tạo kiến trúc model rỗng
    config.ROI_IMG_SIZE = {'HEIGHT': 224, 'WIDTH': 224}
    model_cnn_cmmd_base = create_basic_cnn_model(num_classes=2)

    # 2. Tải trọng số
    model_cnn_cmmd_base.load_weights(MODEL_PATH_CMMD_CNN)
    print(f"\n--- Tải thành công trọng số từ: {MODEL_PATH_CMMD_CNN} ---")

    # 3. Build lại model một cách tường minh để tương thích với Grad-CAM
    print("--- Building model explicitly for Grad-CAM compatibility ---")
    inputs = tf.keras.Input(shape=(config.ROI_IMG_SIZE['HEIGHT'], config.ROI_IMG_SIZE['WIDTH'], 1))
    outputs = model_cnn_cmmd_base(inputs)
    model_cnn_cmmd_built = tf.keras.Model(inputs, outputs, name="Basic_CNN_Built_Correctly")
    
    print("\n--- Model đã được build lại thành công ---")
    model_cnn_cmmd_built.summary()

except Exception as e:
    print(f"\n[LỖI] KHÔNG THỂ TẢI HOẶC BUILD MODEL CNN: {e}")

# =================================================================
# ✅ BƯỚC 2: TẢI ĐƯỜNG DẪN VÀ CHUẨN BỊ DỮ LIỆU (PHIÊN BẢN TỐI ƯU)
# =================================================================

# ✅ Hàm mới: Chỉ lấy đường dẫn file, không đọc ảnh vào bộ nhớ
def get_image_paths_and_labels(data_dir, label_encoder):
    filepaths = []
    labels = []
    print(f"Bắt đầu quét đường dẫn từ: {data_dir}")
    # Lấy tên các lớp (Benign, Malignant) từ label_encoder
    class_names = label_encoder.classes_
    
    for class_name in class_names:
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.isdir(class_dir):
            print(f"[Cảnh báo] Không tìm thấy thư mục lớp: {class_dir}")
            continue
            
        for filename in os.listdir(class_dir):
            # Hỗ trợ các định dạng ảnh phổ biến
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
                filepath = os.path.join(class_dir, filename)
                filepaths.append(filepath)
                labels.append(class_name)
    
    print(f"Đã tìm thấy {len(filepaths)} ảnh.")
    # Chuyển đổi nhãn dạng chữ sang dạng số
    y_numeric = label_encoder.transform(labels)
    return filepaths, y_numeric


# Chỉ thực hiện nếu model đã được tải thành công
if model_cnn_cmmd_built is not None:
    print("\n--- 2. Chuẩn bị đường dẫn dữ liệu CMMD (không tải ảnh) ---")
    DATA_DIR_CMMD = '/kaggle/input/breastdata/CMMD/CMMD'
    le_cmmd = LabelEncoder()
    le_cmmd.fit(['Benign', 'Malignant'])
    
    # ✅ Gọi hàm mới để lấy danh sách đường dẫn
    all_filepaths_cmmd, y_numeric_labels_cmmd = get_image_paths_and_labels(
        data_dir=DATA_DIR_CMMD,
        label_encoder=le_cmmd,
    )
    
    # ✅ Chia train/test trên danh sách ĐƯỜNG DẪN thay vì dữ liệu ảnh
    _, X_test_filepaths, _, y_test_numeric_cmmd = train_test_split(
        all_filepaths_cmmd, y_numeric_labels_cmmd, test_size=0.2, stratify=y_numeric_labels_cmmd, random_state=42
    )
    print(f"Đã chia xong, có {len(X_test_filepaths)} đường dẫn trong tập test CMMD.")

# =================================================================
# BƯỚC 3: ĐỊNH NGHĨA HÀM GRAD-CAM (ĐÃ SỬA LỖI TRUY CẬP LAYER)
# =================================================================
def make_gradcam_heatmap(model, img_array, last_conv_layer_name, pred_index=None):
    # Truy cập vào model con bên trong để lấy layer
    nested_model = model.get_layer('Basic_CNN_Adjusted')
    last_conv_layer = nested_model.get_layer(last_conv_layer_name)

    grad_model = tf.keras.models.Model(
        [model.inputs], [last_conv_layer.output, model.output]
    )
    
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
        
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    
    return tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-8)

def display_gradcam(original_img, heatmap, alpha=0.6, colormap=cv2.COLORMAP_JET):
    # original_img là ảnh float 3 kênh [0,1]
    heatmap_resized = cv2.resize(heatmap.numpy(), (original_img.shape[1], original_img.shape[0]))
    heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap_resized), colormap)
    heatmap_color_rgb = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB).astype(np.float32) / 255.0

    blended_img = (1 - alpha) * original_img + alpha * heatmap_color_rgb
    heatmap_mask = np.expand_dims(heatmap_resized, axis=-1)
    superimposed_img = (original_img * (1 - heatmap_mask) + blended_img * heatmap_mask)
    return np.clip(superimposed_img, 0, 1)

# =================================================================
# BƯỚC 4: TẠO VÀ HIỂN THỊ VISUALIZATION (PHIÊN BẢN TỐI ƯU)
# =================================================================
if model_cnn_cmmd_built is not None and 'X_test_filepaths' in locals():
    print("\n--- 4. Tạo và hiển thị Grad-CAM cho mô hình CNN trên CMMD ---")
    
    LAST_CONV_LAYER_NAME_CNN = "Conv2" 

    # Chọn ảnh mẫu từ tập test
    benign_indices = np.where(y_test_numeric_cmmd == 0)[0]
    malignant_indices = np.where(y_test_numeric_cmmd == 1)[0]
    
    if len(benign_indices) < 2 or len(malignant_indices) < 2:
        print("Cảnh báo: Không đủ ảnh trong tập test để lấy mẫu.")
    else:
        sample_indices = np.concatenate([
            np.random.choice(benign_indices, 2, replace=False),
            np.random.choice(malignant_indices, 2, replace=False)
        ])

        plt.figure(figsize=(12, 12))
        
        for i, img_idx in enumerate(sample_indices):
            # ✅ TẢI ẢNH THEO YÊU CẦU (ON-DEMAND)
            # 1. Lấy đường dẫn của ảnh được chọn
            image_path = X_test_filepaths[img_idx]
            
            # 2. Đọc file ảnh từ đĩa
            original_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            if original_img is None:
                print(f"Lỗi: không thể đọc file ảnh tại {image_path}")
                continue

            # 3. Resize và chuẩn hóa ảnh
            img_resized = cv2.resize(original_img, (config.ROI_IMG_SIZE['WIDTH'], config.ROI_IMG_SIZE['HEIGHT']))
            img_normalized = img_resized.astype(np.float32) / 255.0
            
            # 4. Chuẩn bị ảnh để đưa vào model (thêm chiều batch và channel)
            img_array = np.expand_dims(np.expand_dims(img_normalized, axis=-1), axis=0)

            # Dự đoán
            preds = model_cnn_cmmd_built.predict(img_array)
            predicted_class_idx = np.argmax(preds[0])
            predicted_class_name = le_cmmd.classes_[predicted_class_idx]
            true_class_name = le_cmmd.classes_[y_test_numeric_cmmd[img_idx]]

            # Tạo heatmap
            heatmap = make_gradcam_heatmap(model_cnn_cmmd_built, img_array, LAST_CONV_LAYER_NAME_CNN)
            
            # Chuyển ảnh xám 1 kênh sang RGB 3 kênh để phủ màu
            img_to_display_3channel = cv2.cvtColor(img_normalized, cv2.COLOR_GRAY2RGB)
            superimposed_img = display_gradcam(img_to_display_3channel, heatmap)

            # Hiển thị
            plt.subplot(4, 2, 2*i + 1)
            plt.imshow(img_normalized, cmap='gray') 
            plt.title(f"Original CMMD - True: {true_class_name}")
            plt.axis('off')

            plt.subplot(4, 2, 2*i + 2)
            plt.imshow(superimposed_img)
            plt.title(f"Grad-CAM - Predicted: {predicted_class_name}")
            plt.axis('off')

        plt.tight_layout()
        plt.show()

else:
    if model_cnn_cmmd_built is None:
        print("\n--> Do không tải được model, các bước tạo Grad-CAM đã được bỏ qua.")
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from collections import Counter

# --- CÁC IMPORT TỪ PROJECT CỦA BẠN ---
# Giả định rằng các file/hàm này tồn tại và hoạt động đúng
# Bạn có thể cần tạo một file config.py đơn giản nếu chưa có
# Ví dụ file config.py:
# ROI_IMG_SIZE = {'HEIGHT': 224, 'WIDTH': 224}
import config
from cnn_models.basic_cnn import create_basic_cnn_model

# =================================================================
# BƯỚC 1: CẤU HÌNH VÀ TẢI LẠI MÔ HÌNH CNN ĐÃ HUẤN LUYỆN
# (Giữ nguyên không đổi)
# =================================================================
print("--- 1. Tải lại mô hình CNN đã huấn luyện cho CMMD ---")

MODEL_PATH_CMMD_CNN = '/kaggle/input/cnnelastic1/tensorflow2/default/1/dataset-CMMD_type-all_model-CNN_lr-0.001_b-16_e1-50_e2-50_roi-False_CMMD_CNN_ReducedParams_Elastic1.weights.h5'
model_cnn_cmmd = None
try:
    config.ROI_IMG_SIZE = {'HEIGHT': 224, 'WIDTH': 224}
    model_cnn_cmmd = create_basic_cnn_model(num_classes=2)
    model_cnn_cmmd.load_weights(MODEL_PATH_CMMD_CNN)
    print(f"\n--- Tải thành công model CNN từ: {MODEL_PATH_CMMD_CNN} ---")
    model_cnn_cmmd.summary()

except Exception as e:
    print(f"\n[LỖI] KHÔNG THỂ TẢI MODEL CNN: {e}")
    print("\n==================== HƯỚG DẪN SỬA LỖI ====================")
    print("1. KIỂM TRA LẠI ĐƯỜNG DẪN: Biến 'MODEL_PATH_CMMD_CNN' có chính xác chưa?")
    print("2. KIỂM TRA HÀM 'create_basic_cnn_model': Đảm bảo nó được định nghĩa đúng.")
    print("==========================================================")

# =================================================================
# BƯỚC 2: TẢI VÀ CHUẨN BỊ DỮ LIỆU TEST (PHIÊN BẢN TỐI ƯU)
# =================================================================

# ✅ THAY ĐỔI LỚN: HÀM NÀY CHỈ LẤY ĐƯỜNG DẪN, KHÔNG TẢI ẢNH
def get_cmmd_paths_and_labels(data_dir, label_encoder):
    """
    Quét qua thư mục dataset để lấy danh sách đường dẫn file và nhãn.
    Hàm này không tải dữ liệu ảnh vào bộ nhớ, do đó rất nhẹ.

    Giả định cấu trúc thư mục là:
    - data_dir
        - Benign
            - img1.jpg
            - img2.jpg
        - Malignant
            - img3.jpg
            - ...
    """
    filepaths = []
    labels = []
    print(f"Bắt đầu quét đường dẫn từ: {data_dir}")
    
    # Lấy danh sách các lớp (Benign, Malignant) từ label_encoder
    class_names = label_encoder.classes_
    
    for class_name in class_names:
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.isdir(class_dir):
            print(f"[Cảnh báo] Không tìm thấy thư mục lớp: {class_dir}")
            continue
            
        for filename in os.listdir(class_dir):
            # Hỗ trợ các định dạng ảnh phổ biến
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
                filepath = os.path.join(class_dir, filename)
                filepaths.append(filepath)
                labels.append(class_name)
    
    print(f"Đã tìm thấy {len(filepaths)} ảnh.")
    # Chuyển đổi nhãn dạng chữ sang dạng số
    y_numeric = label_encoder.transform(labels)
    return filepaths, y_numeric

if model_cnn_cmmd is not None:
    print("\n--- 2. Chuẩn bị đường dẫn dữ liệu CMMD (không tải ảnh) ---")
    DATA_DIR_CMMD = '/kaggle/input/breastdata/CMMD/CMMD'
    
    le_cmmd = LabelEncoder()
    le_cmmd.fit(['benign', 'malignant'])
    
    # ✅ SỬ DỤNG HÀM MỚI SIÊU NHẸ
    all_filepaths_cmmd, y_numeric_labels_cmmd = get_cmmd_paths_and_labels(
        data_dir=DATA_DIR_CMMD,
        label_encoder=le_cmmd,
    )
    
    # ✅ Chia train/test trên danh sách ĐƯỜNG DẪN thay vì dữ liệu ảnh
    # Biến X_test_filepaths giờ đây là một list các string, rất nhẹ.
    _, X_test_filepaths, _, y_test_numeric_cmmd = train_test_split(
        all_filepaths_cmmd, y_numeric_labels_cmmd, test_size=0.2, stratify=y_numeric_labels_cmmd, random_state=42
    )
    print(f"Đã chia xong, có {len(X_test_filepaths)} đường dẫn trong tập test CMMD.")

# =================================================================
# BƯỚC 3: ĐỊNH NGHĨA HÀM GRAD-CAM
# (Giữ nguyên không đổi)
# =================================================================
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

def display_gradcam(original_img, heatmap, alpha=0.6, colormap=cv2.COLORMAP_JET):
    heatmap_resized = cv2.resize(heatmap, (original_img.shape[1], original_img.shape[0]))
    heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap_resized), colormap)
    heatmap_color_rgb = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    heatmap_float = heatmap_color_rgb.astype(np.float32) / 255.0

    img_float = original_img.astype(np.float32)
    blended_img = (1 - alpha) * img_float + alpha * heatmap_float
    heatmap_mask = np.expand_dims(heatmap_resized, axis=-1)
    superimposed_img = (img_float * (1 - heatmap_mask) + blended_img * heatmap_mask)
    return np.clip(superimposed_img, 0, 1)

# =================================================================
# BƯỚC 4: TẠO VÀ HIỂN THỊ VISUALIZATION (PHIÊN BẢN TỐI ƯU)
# =================================================================
# ✅ Sửa điều kiện kiểm tra sang `X_test_filepaths`
if model_cnn_cmmd is not None and 'X_test_filepaths' in locals():
    print("\n--- 4. Tạo và hiển thị Grad-CAM cho mô hình CNN trên CMMD ---")
    
    LAST_CONV_LAYER_NAME_CNN = "Conv2"

    # Chọn ngẫu nhiên index từ tập test
    benign_indices = np.where(y_test_numeric_cmmd == 0)[0]
    malignant_indices = np.where(y_test_numeric_cmmd == 1)[0]
    
    # Đảm bảo có đủ ảnh để chọn
    num_samples_per_class = 2
    if len(benign_indices) < num_samples_per_class or len(malignant_indices) < num_samples_per_class:
        print("[Cảnh báo] Không đủ ảnh trong tập test để lấy mẫu. Giảm số lượng mẫu.")
        num_samples_per_class = min(len(benign_indices), len(malignant_indices))

    if num_samples_per_class > 0:
        sample_indices = np.concatenate([
            np.random.choice(benign_indices, num_samples_per_class, replace=False),
            np.random.choice(malignant_indices, num_samples_per_class, replace=False)
        ])

        plt.figure(figsize=(12, 12))
        
        for i, img_idx in enumerate(sample_indices):
            # ✅ THAY ĐỔI: TẢI VÀ XỬ LÝ CHỈ 1 ẢNH TẠI ĐÂY
            # 1. Lấy đường dẫn của ảnh được chọn
            image_path = X_test_filepaths[img_idx]
            
            # 2. Đọc file ảnh từ đĩa
            img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            
            # 3. Resize và chuẩn hóa ảnh (giống như khi train)
            img_resized = cv2.resize(img, (config.ROI_IMG_SIZE['WIDTH'], config.ROI_IMG_SIZE['HEIGHT']))
            img_normalized = img_resized.astype(np.float32) / 255.0
            
            # 4. Thêm chiều kênh (channel) để phù hợp với input của model: (224, 224) -> (224, 224, 1)
            original_img_with_channel = np.expand_dims(img_normalized, axis=-1)
            
            # 5. Thêm chiều batch để đưa vào model: (224, 224, 1) -> (1, 224, 224, 1)
            img_array = np.expand_dims(original_img_with_channel, axis=0)

            # ----- Các bước còn lại giữ nguyên logic -----
            
            # Dự đoán
            preds = model_cnn_cmmd.predict(img_array)
            predicted_class_idx = np.argmax(preds[0])
            predicted_class_name = le_cmmd.classes_[predicted_class_idx]
            true_class_name = le_cmmd.classes_[y_test_numeric_cmmd[img_idx]]

            # Tạo heatmap
            heatmap = make_gradcam_heatmap(img_array, model_cnn_cmmd, LAST_CONV_LAYER_NAME_CNN)
            
            # Chuyển ảnh xám 1 kênh sang 3 kênh để phủ màu Grad-CAM
            img_to_display_3channel = cv2.cvtColor(img_normalized, cv2.COLOR_GRAY2RGB)
            superimposed_img = display_gradcam(img_to_display_3channel, heatmap, alpha=0.5)

            # Hiển thị ảnh gốc
            plt.subplot(4, 2, 2 * i + 1)
            plt.imshow(img_normalized, cmap='gray') # Hiển thị ảnh đã chuẩn hóa
            plt.title(f"Original CMMD - True: {true_class_name}")
            plt.axis('off')

            # Hiển thị ảnh với Grad-CAM
            plt.subplot(4, 2, 2 * i + 2)
            plt.imshow(superimposed_img)
            plt.title(f"Grad-CAM - Predicted: {predicted_class_name}")
            plt.axis('off')

        plt.tight_layout()
        plt.show()

else:
    if model_cnn_cmmd is None:
        print("\n--> Do không tải được model, các bước tạo Grad-CAM đã được bỏ qua.")
    elif 'X_test_filepaths' not in locals():
        print("\n--> Không tìm thấy đường dẫn dữ liệu test, các bước tạo Grad-CAM đã được bỏ qua.")
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from collections import Counter

# --- CÁC IMPORT TỪ PROJECT CỦA BẠN ---
# Giả định rằng các file/hàm này tồn tại và hoạt động đúng
# Bạn có thể cần tạo một file config.py đơn giản nếu chưa có
# Ví dụ file config.py:
# ROI_IMG_SIZE = {'HEIGHT': 224, 'WIDTH': 224}
import config
from cnn_models.basic_cnn import create_basic_cnn_model

# =================================================================
# BƯỚC 1: CẤU HÌNH VÀ TẢI LẠI MÔ HÌNH CNN ĐÃ HUẤN LUYỆN
# (Giữ nguyên không đổi)
# =================================================================
print("--- 1. Tải lại mô hình CNN đã huấn luyện cho CMMD ---")

MODEL_PATH_CMMD_CNN = '/kaggle/input/cnnelastic1/tensorflow2/default/1/dataset-CMMD_type-all_model-CNN_lr-0.001_b-16_e1-50_e2-50_roi-False_CMMD_CNN_ReducedParams_Elastic1.weights.h5'

model_cnn_cmmd = None
try:
    config.ROI_IMG_SIZE = {'HEIGHT': 224, 'WIDTH': 224}
    model_cnn_cmmd = create_basic_cnn_model(num_classes=2)
    model_cnn_cmmd.load_weights(MODEL_PATH_CMMD_CNN)
    print(f"\n--- Tải thành công model CNN từ: {MODEL_PATH_CMMD_CNN} ---")
    model_cnn_cmmd.summary()

except Exception as e:
    print(f"\n[LỖI] KHÔNG THỂ TẢI MODEL CNN: {e}")
    print("\n==================== HƯỚG DẪN SỬA LỖI ====================")
    print("1. KIỂM TRA LẠI ĐƯỜNG DẪN: Biến 'MODEL_PATH_CMMD_CNN' có chính xác chưa?")
    print("2. KIỂM TRA HÀM 'create_basic_cnn_model': Đảm bảo nó được định nghĩa đúng.")
    print("==========================================================")

# =================================================================
# BƯỚC 2: TẢI VÀ CHUẨN BỊ DỮ LIỆU TEST (PHIÊN BẢN TỐI ƯU)
# =================================================================

# ✅ THAY ĐỔI LỚN: HÀM NÀY CHỈ LẤY ĐƯỜNG DẪN, KHÔNG TẢI ẢNH
def get_cmmd_paths_and_labels(data_dir, label_encoder):
    """
    Quét qua thư mục dataset để lấy danh sách đường dẫn file và nhãn.
    Hàm này không tải dữ liệu ảnh vào bộ nhớ, do đó rất nhẹ.

    Giả định cấu trúc thư mục là:
    - data_dir
        - Benign
            - img1.jpg
            - img2.jpg
        - Malignant
            - img3.jpg
            - ...
    """
    filepaths = []
    labels = []
    print(f"Bắt đầu quét đường dẫn từ: {data_dir}")
    
    # Lấy danh sách các lớp (Benign, Malignant) từ label_encoder
    class_names = label_encoder.classes_
    
    for class_name in class_names:
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.isdir(class_dir):
            print(f"[Cảnh báo] Không tìm thấy thư mục lớp: {class_dir}")
            continue
            
        for filename in os.listdir(class_dir):
            # Hỗ trợ các định dạng ảnh phổ biến
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
                filepath = os.path.join(class_dir, filename)
                filepaths.append(filepath)
                labels.append(class_name)
    
    print(f"Đã tìm thấy {len(filepaths)} ảnh.")
    # Chuyển đổi nhãn dạng chữ sang dạng số
    y_numeric = label_encoder.transform(labels)
    return filepaths, y_numeric

if model_cnn_cmmd is not None:
    print("\n--- 2. Chuẩn bị đường dẫn dữ liệu CMMD (không tải ảnh) ---")
    DATA_DIR_CMMD = '/kaggle/input/breastdata/CMMD/CMMD'
    
    le_cmmd = LabelEncoder()
    le_cmmd.fit(['benign', 'malignant'])
    
    # ✅ SỬ DỤNG HÀM MỚI SIÊU NHẸ
    all_filepaths_cmmd, y_numeric_labels_cmmd = get_cmmd_paths_and_labels(
        data_dir=DATA_DIR_CMMD,
        label_encoder=le_cmmd,
    )
    
    # ✅ Chia train/test trên danh sách ĐƯỜNG DẪN thay vì dữ liệu ảnh
    # Biến X_test_filepaths giờ đây là một list các string, rất nhẹ.
    _, X_test_filepaths, _, y_test_numeric_cmmd = train_test_split(
        all_filepaths_cmmd, y_numeric_labels_cmmd, test_size=0.2, stratify=y_numeric_labels_cmmd, random_state=42
    )
    print(f"Đã chia xong, có {len(X_test_filepaths)} đường dẫn trong tập test CMMD.")

# =================================================================
# BƯỚC 3: ĐỊNH NGHĨA HÀM GRAD-CAM
# (Giữ nguyên không đổi)
# =================================================================
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

def display_gradcam(original_img, heatmap, alpha=0.6, colormap=cv2.COLORMAP_JET):
    heatmap_resized = cv2.resize(heatmap, (original_img.shape[1], original_img.shape[0]))
    heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap_resized), colormap)
    heatmap_color_rgb = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    heatmap_float = heatmap_color_rgb.astype(np.float32) / 255.0

    img_float = original_img.astype(np.float32)
    blended_img = (1 - alpha) * img_float + alpha * heatmap_float
    heatmap_mask = np.expand_dims(heatmap_resized, axis=-1)
    superimposed_img = (img_float * (1 - heatmap_mask) + blended_img * heatmap_mask)
    return np.clip(superimposed_img, 0, 1)

# =================================================================
# BƯỚC 4: TẠO VÀ HIỂN THỊ VISUALIZATION (PHIÊN BẢN TỐI ƯU)
# =================================================================
# ✅ Sửa điều kiện kiểm tra sang `X_test_filepaths`
if model_cnn_cmmd is not None and 'X_test_filepaths' in locals():
    print("\n--- 4. Tạo và hiển thị Grad-CAM cho mô hình CNN trên CMMD ---")
    
    LAST_CONV_LAYER_NAME_CNN = "Conv2"

    # Chọn ngẫu nhiên index từ tập test
    benign_indices = np.where(y_test_numeric_cmmd == 0)[0]
    malignant_indices = np.where(y_test_numeric_cmmd == 1)[0]
    
    # Đảm bảo có đủ ảnh để chọn
    num_samples_per_class = 2
    if len(benign_indices) < num_samples_per_class or len(malignant_indices) < num_samples_per_class:
        print("[Cảnh báo] Không đủ ảnh trong tập test để lấy mẫu. Giảm số lượng mẫu.")
        num_samples_per_class = min(len(benign_indices), len(malignant_indices))

    if num_samples_per_class > 0:
        sample_indices = np.concatenate([
            np.random.choice(benign_indices, num_samples_per_class, replace=False),
            np.random.choice(malignant_indices, num_samples_per_class, replace=False)
        ])

        plt.figure(figsize=(12, 12))
        
        for i, img_idx in enumerate(sample_indices):
            # ✅ THAY ĐỔI: TẢI VÀ XỬ LÝ CHỈ 1 ẢNH TẠI ĐÂY
            # 1. Lấy đường dẫn của ảnh được chọn
            image_path = X_test_filepaths[img_idx]
            
            # 2. Đọc file ảnh từ đĩa
            img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            
            # 3. Resize và chuẩn hóa ảnh (giống như khi train)
            img_resized = cv2.resize(img, (config.ROI_IMG_SIZE['WIDTH'], config.ROI_IMG_SIZE['HEIGHT']))
            img_normalized = img_resized.astype(np.float32) / 255.0
            
            # 4. Thêm chiều kênh (channel) để phù hợp với input của model: (224, 224) -> (224, 224, 1)
            original_img_with_channel = np.expand_dims(img_normalized, axis=-1)
            
            # 5. Thêm chiều batch để đưa vào model: (224, 224, 1) -> (1, 224, 224, 1)
            img_array = np.expand_dims(original_img_with_channel, axis=0)

            # ----- Các bước còn lại giữ nguyên logic -----
            
            # Dự đoán
            preds = model_cnn_cmmd.predict(img_array)
            predicted_class_idx = np.argmax(preds[0])
            predicted_class_name = le_cmmd.classes_[predicted_class_idx]
            true_class_name = le_cmmd.classes_[y_test_numeric_cmmd[img_idx]]

            # Tạo heatmap
            heatmap = make_gradcam_heatmap(img_array, model_cnn_cmmd, LAST_CONV_LAYER_NAME_CNN)
            
            # Chuyển ảnh xám 1 kênh sang 3 kênh để phủ màu Grad-CAM
            img_to_display_3channel = cv2.cvtColor(img_normalized, cv2.COLOR_GRAY2RGB)
            superimposed_img = display_gradcam(img_to_display_3channel, heatmap, alpha=0.5)

            # Hiển thị ảnh gốc
            plt.subplot(4, 2, 2 * i + 1)
            plt.imshow(img_normalized, cmap='gray') # Hiển thị ảnh đã chuẩn hóa
            plt.title(f"Original CMMD - True: {true_class_name}")
            plt.axis('off')

            # Hiển thị ảnh với Grad-CAM
            plt.subplot(4, 2, 2 * i + 2)
            plt.imshow(superimposed_img)
            plt.title(f"Grad-CAM - Predicted: {predicted_class_name}")
            plt.axis('off')

        plt.tight_layout()
        plt.show()

else:
    if model_cnn_cmmd is None:
        print("\n--> Do không tải được model, các bước tạo Grad-CAM đã được bỏ qua.")
    elif 'X_test_filepaths' not in locals():
        print("\n--> Không tìm thấy đường dẫn dữ liệu test, các bước tạo Grad-CAM đã được bỏ qua.")
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from collections import Counter

# --- CÁC IMPORT TỪ PROJECT CỦA BẠN ---
# Giả sử các file/hàm này tồn tại và hoạt động đúng
import config
from cnn_models.basic_cnn import create_basic_cnn_model

# =================================================================
# BƯỚC 1: CẤU HÌNH, TẢI LẠI VÀ BUILD LẠI MÔ HÌNH
# =================================================================
print("--- 1. Tải lại mô hình CNN đã huấn luyện cho CMMD ---")

MODEL_PATH_CMMD_CNN = '/kaggle/input/cnnelastic1/tensorflow2/default/1/dataset-CMMD_type-all_model-CNN_lr-0.001_b-16_e1-50_e2-50_roi-False_CMMD_CNN_ReducedParams_Elastic1.weights.h5'

# ✅ ĐỔI TÊN BIẾN ĐỂ TRÁNH NHẦM LẪN
model_cnn_cmmd_base = None
model_cnn_cmmd_built = None # Model sẽ được build lại và sử dụng

try:
    config.ROI_IMG_SIZE = {'HEIGHT': 224, 'WIDTH': 224}
    # Tải kiến trúc và trọng số vào model gốc
    model_cnn_cmmd_base = create_basic_cnn_model(num_classes=2)
    model_cnn_cmmd_base.load_weights(MODEL_PATH_CMMD_CNN)
    print(f"\n--- Tải thành công model gốc từ: {MODEL_PATH_CMMD_CNN} ---")

    # ✅ BƯỚC QUAN TRỌNG: BUILD LẠI MODEL MỘT CÁCH TƯỜNG MINH
    print("--- Building model explicitly for Grad-CAM compatibility ---")
    # 1. Tạo một input tensor với shape chính xác
    inputs = tf.keras.Input(shape=(config.ROI_IMG_SIZE['HEIGHT'], config.ROI_IMG_SIZE['WIDTH'], 1))
    # 2. Gọi model gốc trên input mới để lấy output
    outputs = model_cnn_cmmd_base(inputs)
    # 3. Tạo một model mới đã được build đúng cách
    model_cnn_cmmd_built = tf.keras.Model(inputs, outputs, name="Basic_CNN_Built_Correctly")
    
    print("\n--- Model đã được build lại thành công ---")
    model_cnn_cmmd_built.summary()

except Exception as e:
    print(f"\n[LỖI] KHÔNG THỂ TẢI HOẶC BUILD MODEL CNN: {e}")
    print("\n==================== HƯỚG DẪN SỬA LỖI ====================")
    print("1. KIỂM TRA LẠI ĐƯỜNG DẪN: Biến 'MODEL_PATH_CMMD_CNN' có chính xác chưa?")
    print("2. KIỂM TRA HÀM 'create_basic_cnn_model': Đảm bảo nó được định nghĩa đúng.")
    print("==========================================================")


# =================================================================
# BƯỚC 2: TẢI VÀ CHUẨN BỊ DỮ LIỆU TEST (PHIÊN BẢN TỐI ƯU)
# (Giữ nguyên không đổi)
# =================================================================
def get_cmmd_paths_and_labels(data_dir, label_encoder):
    filepaths = []
    labels = []
    print(f"Bắt đầu quét đường dẫn từ: {data_dir}")
    class_names = label_encoder.classes_
    
    for class_name in class_names:
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.isdir(class_dir):
            print(f"[Cảnh báo] Không tìm thấy thư mục lớp '{class_name}' tại: {class_dir}")
            continue
            
        for filename in os.listdir(class_dir):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
                filepath = os.path.join(class_dir, filename)
                filepaths.append(filepath)
                labels.append(class_name)
    
    print(f"Đã tìm thấy {len(filepaths)} ảnh.")
    y_numeric = label_encoder.transform(labels)
    return filepaths, y_numeric

# ✅ Sửa điều kiện kiểm tra sang model đã được build
if model_cnn_cmmd_built is not None:
    print("\n--- 2. Chuẩn bị đường dẫn dữ liệu CMMD (không tải ảnh) ---")
    DATA_DIR_CMMD = '/kaggle/input/breastdata/CMMD/CMMD'
    
    le_cmmd = LabelEncoder()
    # ✅ Sửa lại nhãn để khớp với tên thư mục (thường là viết thường)
    le_cmmd.fit(['benign', 'malignant'])
    
    all_filepaths_cmmd, y_numeric_labels_cmmd = get_cmmd_paths_and_labels(
        data_dir=DATA_DIR_CMMD,
        label_encoder=le_cmmd,
    )
    
    _, X_test_filepaths, _, y_test_numeric_cmmd = train_test_split(
        all_filepaths_cmmd, y_numeric_labels_cmmd, test_size=0.2, stratify=y_numeric_labels_cmmd, random_state=42
    )
    print(f"Đã chia xong, có {len(X_test_filepaths)} đường dẫn trong tập test CMMD.")


# =================================================================
# BƯỚC 3: ĐỊNH NGHĨA HÀM GRAD-CAM
# (Giữ nguyên không đổi)
# =================================================================
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    # Thêm hằng số nhỏ để tránh chia cho 0
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-8)
    return heatmap.numpy()

def display_gradcam(original_img, heatmap, alpha=0.6, colormap=cv2.COLORMAP_JET):
    heatmap_resized = cv2.resize(heatmap, (original_img.shape[1], original_img.shape[0]))
    heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap_resized), colormap)
    heatmap_color_rgb = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    heatmap_float = heatmap_color_rgb.astype(np.float32) / 255.0

    img_float = original_img.astype(np.float32)
    blended_img = (1 - alpha) * img_float + alpha * heatmap_float
    heatmap_mask = np.expand_dims(heatmap_resized, axis=-1)
    superimposed_img = (img_float * (1 - heatmap_mask) + blended_img * heatmap_mask)
    return np.clip(superimposed_img, 0, 1)

# =================================================================
# BƯỚC 4: TẠO VÀ HIỂN THỊ VISUALIZATION (PHIÊN BẢN TỐI ƯU)
# =================================================================
# ✅ Sửa điều kiện kiểm tra và các lời gọi hàm để dùng model_cnn_cmmd_built
if model_cnn_cmmd_built is not None and 'X_test_filepaths' in locals() and len(X_test_filepaths) > 0:
    print("\n--- 4. Tạo và hiển thị Grad-CAM cho mô hình CNN trên CMMD ---")
    
    # ✅ Đảm bảo tên lớp Conv2 trong model của bạn là "Conv2"
    # Bạn có thể kiểm tra lại từ model.summary() ở trên
    LAST_CONV_LAYER_NAME_CNN = "Conv2" 

    benign_indices = np.where(y_test_numeric_cmmd == 0)[0]
    malignant_indices = np.where(y_test_numeric_cmmd == 1)[0]
    
    num_samples_per_class = 2
    if len(benign_indices) < num_samples_per_class or len(malignant_indices) < num_samples_per_class:
        print("[Cảnh báo] Không đủ ảnh trong tập test để lấy mẫu. Giảm số lượng mẫu.")
        num_samples_per_class = min(len(benign_indices), len(malignant_indices))

    if num_samples_per_class > 0:
        sample_indices = np.concatenate([
            np.random.choice(benign_indices, num_samples_per_class, replace=False),
            np.random.choice(malignant_indices, num_samples_per_class, replace=False)
        ])

        plt.figure(figsize=(12, 12))
        
        for i, img_idx in enumerate(sample_indices):
            image_path = X_test_filepaths[img_idx]
            img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            
            # Kiểm tra nếu ảnh không đọc được
            if img is None:
                print(f"[Lỗi] Không thể đọc ảnh tại: {image_path}")
                continue

            img_resized = cv2.resize(img, (config.ROI_IMG_SIZE['WIDTH'], config.ROI_IMG_SIZE['HEIGHT']))
            img_normalized = img_resized.astype(np.float32) / 255.0
            original_img_with_channel = np.expand_dims(img_normalized, axis=-1)
            img_array = np.expand_dims(original_img_with_channel, axis=0)

            # ✅ SỬ DỤNG MODEL ĐÃ ĐƯỢC BUILD LẠI
            preds = model_cnn_cmmd_built.predict(img_array)
            predicted_class_idx = np.argmax(preds[0])
            predicted_class_name = le_cmmd.classes_[predicted_class_idx]
            true_class_name = le_cmmd.classes_[y_test_numeric_cmmd[img_idx]]

            # ✅ SỬ DỤNG MODEL ĐÃ ĐƯỢC BUILD LẠI
            heatmap = make_gradcam_heatmap(img_array, model_cnn_cmmd_built, LAST_CONV_LAYER_NAME_CNN)
            
            img_to_display_3channel = cv2.cvtColor(img_normalized, cv2.COLOR_GRAY2RGB)
            superimposed_img = display_gradcam(img_to_display_3channel, heatmap, alpha=0.5)

            plt.subplot(4, 2, 2 * i + 1)
            plt.imshow(img_normalized, cmap='gray')
            plt.title(f"Original CMMD - True: {true_class_name}")
            plt.axis('off')

            plt.subplot(4, 2, 2 * i + 2)
            plt.imshow(superimposed_img)
            plt.title(f"Grad-CAM - Predicted: {predicted_class_name}")
            plt.axis('off')

        plt.tight_layout()
        plt.show()

else:
    if model_cnn_cmmd_built is None:
        print("\n--> Do không tải hoặc build được model, các bước tạo Grad-CAM đã được bỏ qua.")
    elif 'X_test_filepaths' not in locals() or len(X_test_filepaths) == 0:
        print("\n--> Không tìm thấy đường dẫn dữ liệu test, các bước tạo Grad-CAM đã được bỏ qua.")
```

```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# --- CÁC IMPORT TỪ PROJECT CỦA BẠN ---
# Giả sử các hàm này đã được định nghĩa đúng
# Bạn có thể cần tạo một file config.py đơn giản nếu chưa có
# Ví dụ file config.py:
# ROI_IMG_SIZE = {'HEIGHT': 224, 'WIDTH': 224}
# dataset = "CMMD"
# model = "MobileNet"
# verbose_mode = False
import config
from cnn_models.mobilenet_v2 import create_mobilenet_model


# =================================================================
# HÀM GRAD-CAM VÀ HIỂN THỊ
# (Giữ nguyên, không cần thay đổi)
# =================================================================
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-8)
    return heatmap.numpy()

def display_gradcam(original_img_rgb, heatmap, alpha=0.6, colormap=cv2.COLORMAP_JET):
    heatmap_resized = cv2.resize(heatmap, (original_img_rgb.shape[1], original_img_rgb.shape[0]))
    heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap_resized), colormap)
    heatmap_color_rgb = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    
    heatmap_float = heatmap_color_rgb.astype(np.float32) / 255.0
    img_float = original_img_rgb.astype(np.float32) # Ảnh gốc đã là float và RGB
    
    blended_img = (1 - alpha) * img_float + alpha * heatmap_float
    return np.clip(blended_img, 0, 1)

# ✅ HÀM LẤY ĐƯỜNG DẪN SIÊU NHẸ
def get_cmmd_paths_and_labels(data_dir, label_encoder):
    filepaths = []
    labels = []
    print(f"Bắt đầu quét đường dẫn từ: {data_dir}")
    class_names = label_encoder.classes_
    
    for class_name in class_names:
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.isdir(class_dir):
            print(f"[Cảnh báo] Không tìm thấy thư mục lớp: {class_dir}")
            continue
            
        for filename in os.listdir(class_dir):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
                filepath = os.path.join(class_dir, filename)
                filepaths.append(filepath)
                labels.append(class_name)
    
    print(f"Đã tìm thấy {len(filepaths)} ảnh.")
    y_numeric = label_encoder.transform(labels)
    return filepaths, y_numeric

# =================================================================
# THIẾT LẬP THAM SỐ VÀ CHẠY
# =================================================================

MODEL_PATH = '/kaggle/input/mobilenetelastic2/tensorflow2/default/1'
DATA_DIR = '/kaggle/input/breastdata/CMMD/CMMD'

print("--- Bắt đầu quy trình Grad-CAM cho MobileNet ---")

# --- 1. Tải lại mô hình MobileNet ---
print("--- 1. Tải lại mô hình MobileNet đã huấn luyện ---")
model_cmmd = None
if not os.path.exists(MODEL_PATH) or not os.path.isdir(MODEL_PATH):
    print(f"[LỖI] Không tìm thấy thư mục model tại: {MODEL_PATH}")
else:
    try:
        config.dataset = "CMMD"
        config.model = "MobileNet"
        config.ROI_IMG_SIZE = {'HEIGHT': 224, 'WIDTH': 224} # Đảm bảo config này có
        
        model_cmmd = create_mobilenet_model(num_classes=2)
        model_cmmd.load_weights(MODEL_PATH)
        print(f"\n--- Tải thành công model MobileNet từ: {MODEL_PATH} ---")
        model_cmmd.summary()
    except Exception as e:
        print(f"\n[LỖI] KHÔNG THỂ TẢI MODEL MOBILENET: {e}")

if model_cmmd is None:
    print("Không thể tiếp tục vì không tải được mô hình.")
else:
    # --- 2. Chuẩn bị dữ liệu TEST (phiên bản tối ưu) ---
    print("\n--- 2. Chuẩn bị đường dẫn dữ liệu CMMD để visualize ---")
    if not os.path.isdir(DATA_DIR):
        print(f"[LỖI] Không tìm thấy thư mục dữ liệu tại: {DATA_DIR}")
    else:
        le_cmmd = LabelEncoder()
        le_cmmd.fit(['benign', 'malignant'])
        
        # ✅ SỬ DỤNG HÀM MỚI SIÊU NHẸ
        all_filepaths_cmmd, y_numeric_labels_cmmd = get_cmmd_paths_and_labels(
            data_dir=DATA_DIR, label_encoder=le_cmmd
        )
        
        # ✅ Chia train/test trên danh sách ĐƯỜNG DẪN
        _, X_test_filepaths, _, y_test_numeric_cmmd = train_test_split(
            all_filepaths_cmmd, y_numeric_labels_cmmd, test_size=0.2, stratify=y_numeric_labels_cmmd, random_state=42
        )
        print(f"Đã chia xong, có {len(X_test_filepaths)} đường dẫn trong tập test.")

        # --- 3. Tạo và hiển thị Grad-CAM (phiên bản tối ưu cho MobileNet) ---
        print("\n--- 3. Tạo và hiển thị Grad-CAM ---")
        # ✅ Lớp conv cuối cùng của MobileNetV2 là 'out_relu'
        LAST_CONV_LAYER_NAME = "out_relu"

        benign_indices = np.where(y_test_numeric_cmmd == 0)[0]
        malignant_indices = np.where(y_test_numeric_cmmd == 1)[0]
        
        num_samples_per_class = 2
        if len(benign_indices) < num_samples_per_class or len(malignant_indices) < num_samples_per_class:
            print("[Cảnh báo] Không đủ mẫu trong tập test để visualize.")
        else:
            sample_indices = np.concatenate([
                np.random.choice(benign_indices, num_samples_per_class, replace=False),
                np.random.choice(malignant_indices, num_samples_per_class, replace=False)
            ])

            plt.figure(figsize=(15, 5 * len(sample_indices)))
            
            for i, img_idx in enumerate(sample_indices):
                # ✅ THAY ĐỔI: Tải và xử lý CHỈ 1 ẢNH tại đây
                image_path = X_test_filepaths[img_idx]
                img_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
                img_gray_resized = cv2.resize(img_gray, (config.ROI_IMG_SIZE['WIDTH'], config.ROI_IMG_SIZE['HEIGHT']))
                img_gray_normalized = img_gray_resized.astype(np.float32) / 255.0
                
                # ✅ QUAN TRỌNG: Chuyển ảnh xám 1 kênh sang RGB 3 kênh cho MobileNet
                img_rgb = cv2.cvtColor(img_gray_normalized, cv2.COLOR_GRAY2RGB)
                
                # Tạo batch size 1 để đưa vào model: (H, W, 3) -> (1, H, W, 3)
                img_array_for_model = np.expand_dims(img_rgb, axis=0)
                
                # Dự đoán lớp
                preds = model_cmmd.predict(img_array_for_model)
                predicted_class_idx = np.argmax(preds[0])
                predicted_class_name = le_cmmd.classes_[predicted_class_idx]
                true_class_name = le_cmmd.classes_[y_test_numeric_cmmd[img_idx]]

                # Tạo heatmap Grad-CAM
                heatmap = make_gradcam_heatmap(img_array_for_model, model_cmmd, LAST_CONV_LAYER_NAME)
                
                # Phủ heatmap lên ảnh RGB gốc
                superimposed_img = display_gradcam(img_rgb, heatmap, alpha=0.5)

                # Hiển thị kết quả
                plt.subplot(len(sample_indices), 3, 3*i + 1)
                plt.imshow(img_rgb)
                plt.title(f"Original (RGB) - True: {true_class_name}")
                plt.axis('off')

                plt.subplot(len(sample_indices), 3, 3*i + 2)
                plt.imshow(heatmap, cmap='viridis')
                plt.title("Raw Heatmap")
                plt.axis('off')

                plt.subplot(len(sample_indices), 3, 3*i + 3)
                plt.imshow(superimposed_img)
                plt.title(f"Grad-CAM - Predicted: {predicted_class_name}")
                plt.axis('off')

            plt.tight_layout()
            plt.show()

print("\n--- Quy trình Grad-CAM hoàn tất ---")
```
